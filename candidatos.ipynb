## Simulação Eleições
<hr>

Gerem uma população com 10 mil "eleitores": 3000 votam no candidato A e 7000 no candidato B. A proporção populacional de votos no candidato A é portanto p=0.30.

Siga os passos:
<ol>
<li> realize uma amostra aleatória (com ou sem reposição) de tamanho n.</li>

<li> calcule a proporção amostral de votos no candidato A: "p chapéu".</li>

<li> repita os passos 1) e 2) 2000 vezes.</li>

<li> faça um gráfico no valores obtidos de "p chapéu": histograma ou gráfico como feito na aula.</li>

<li> calcule o média dos valores obtidos de "p chapéu".</li>

<li> calcule o desvio padrão dos valores obtidos de "p chapéu".</li>
</ol>
Faça isso para 2 tamanhos diferentes de amostra: n (fica a seu critério escolher dois tamanhos amostrais bem diferentes).

Comente os resultados obtidos com os tamanhos amostrais diferentes, em termo de precisão e proximidade à proporção populacional.

```python
import random as rd
from statistics import mean, stdev
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### Constantes do problema
<hr>

Foram escolhidos n<sub>1</sub> = 500 e n<sub>2</sub> = 100

```python
N = int(1e4) # population size
p = .3 # population probability
n = int(.05*N), int(.01*N) # sample sizes
k = 2000 # n of samples taken
```

### Cálculo de p̂
<hr>

A população é um vetor de 3000 valores = 'a' e 7000 valores = 'b'<br>
p̂ é um vetor 2xk com cada valor para a probabilidade amostral em que cada linha corresponde a um valor de n

```python
pop = ['a' if i < int(p*N) else 'b' for i in range(N)]
p̂ = [[rd.sample(pop, n[j]).count('a')/n[j] for i in range(k)] for j in range(len(n))]
```

### Resultados
<hr>

```python
plt.figure(figsize=(45, 15))
colors = ['m', 'g', 'k', 'b', 'r', 'c']
for i in range(len(n)):
    ax = plt.subplot(211+i)
    ax.plot(p̂[i], color=colors[i])
    ax.set_title('Trial {}'.format(i+1), fontsize=36)
    ax.set_xbound(0, k)
    ax.set_xlabel('Sample', fontsize=36)
    ax.set_ylabel('p̂', fontsize=36)
plt.subplots_adjust(hspace=.5)
mpl.rcParams.update({'font.size': 30})
plt.show()
for i in range(len(n)):
    print("Trial {}:".format(i+1))
    print("\tAverage: {0:.5f}\n\tStandard Deviation: {1:.5f}\n".format(mean(p̂[i]), stdev(p̂[i])))
```

### Conclusões
<hr>

Para um n maior, p̂ é na média mais próximo de p com um desvio padrão menor.
